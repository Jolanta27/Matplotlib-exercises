1. Create a chart:

select the last 30 observations from the weather collection
set a size of the graph
create a line graph for the variables wind and precipitation
change the marker type, color, and line type
add a legend with line description

```python
url='https://raw.githubusercontent.com/matzim95/ML-datasets/master/weather.csv'

weather=pd.read_csv(
    url,
    parse_dates=True,
    index_col='date'
)
weather.head()
```

```python
n = 30
weather.index[:n]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(weather.index, weather['wind'], marker='o', color='b', linestyle='-', label='Wind')
plt.plot(weather.index, weather['precipitation'], marker='x', color='r', linestyle='--', label='Precipitation')

plt.legend()

plt.show()
```

```python
n=30

fig,ax = plt.subplots(figsize=(8, 4))

ax.plot(
    weather.index[-n:],
    weather['temp_max'][-n:],
    marker='s',
    linestyle='solid',
    color='maroon',
    label='max'
)

ax.plot(
    weather.index[-n:],
    weather['temp_min'][-n:],
    marker='d',
    linestyle='dotted',
    color='forestgreen',
    label='min'
)

font={'family':'serif',
      'color':'navy',
      'weight':'normal',
      'size':16}

#pad - podniesienie/obnienie tytuu wykresu
ax.set_title('Temperature in Seattle - December 2015',fontdict=font,pad=10)  # plt.title()
#labelpad - odlego tytuu osi od osi
ax.set_xlabel('Date (year/month/day)',labelpad=10,fontdict=font)  # plt.xlabel()
ax.set_ylabel('Temperature [centigrade]',fontdict=font)  # plt.ylabel()
# ax.xaxis.set_tick_params(rotation=90)  # eq. to ax.xticks()

ax.legend(title='Temperature')
plt.xticks(rotation=270)
plt.show()
```

Create graph based on the weather dataset for Austin:

for the fourth quarter of 2015
include temperature (avg, high, low), humidity (avg, high, low) and wind (avg, high, gust) on the graph
add annotation to the graph, ie. mark when the strongest windstorm was in the period under study


```python
url='https://raw.githubusercontent.com/matzim95/ML-datasets/master/austin_weather.csv'

austin_weather = pd.read_csv(url, parse_dates=True, index_col='Date')
austin_weather.head()
```

```python
import matplotlib.pyplot as plt

# Filter the data for the fourth quarter of 2015
weather_q4_2015 = weather['2015-10-01':'2015-12-31']

# Create a new figure and set its size
plt.figure(figsize=(10, 6))

# Plot the temperature, humidity, and wind data
plt.plot(weather_q4_2015.index, weather_q4_2015['temp_max'], label='Max Temperature')
plt.plot(weather_q4_2015.index, weather_q4_2015['temp_min'], label='Min Temperature')
plt.plot(weather_q4_2015.index, weather_q4_2015['wind'], label='Wind')


strongest_wind_date = weather_q4_2015['wind'].idxmax()

# Add an annotation for the strongest wind
plt.annotate('Strongest wind', xy=(strongest_wind_date, weather_q4_2015.loc[strongest_wind_date, 'wind']), xytext=(strongest_wind_date, weather_q4_2015.loc[strongest_wind_date, 'wind'] + 1), arrowprops=dict(facecolor='black', shrink=0.05))

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure the index is a datetime index
weather.index = pd.to_datetime(weather.index)

# Filter the data for the year 2014
weather_2014 = weather.loc['2014']

# Create a new figure and set its size
fig, ax = plt.subplots(2, 1, figsize=(10, 6))

# Create a graph for precipitation
ax[0].plot(weather_2014.index, weather_2014['precipitation'], color='blue', marker='o', linestyle='-')
ax[0].set_title('Precipitation in 2014')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Precipitation')

# Create a graph for wind strength
ax[1].plot(weather_2014.index, weather_2014['wind'], color='red', marker='x', linestyle='--')
ax[1].set_title('Wind Strength in 2014')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Wind Strength')

# Customize the aesthetics of the graph
plt.tight_layout()
plt.show()
```
