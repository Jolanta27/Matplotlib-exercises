```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd

def create_scatter_plot(data=None, x=None, y=None, color=None):
    if data is not None:
        x = data[x]
        y = data[y]
        color = data[color]

    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, c=color)
    plt.colorbar()
    plt.show()

# Assuming df is your DataFrame
df = pd.DataFrame({
    'temp_max': [12.8, 10.6, 11.7, 12.2, 8.9],
    'temp_min': [5.0, 2.8, 7.2, 5.6, 2.8],
    'precipitation': [0.0, 10.9, 0.8, 20.3, 1.3]
})

create_scatter_plot(data=df, x='temp_max', y='temp_min', color='precipitation')
```

```python
url='https://raw.githubusercontent.com/matzim95/ML-datasets/master/austin_weather.csv'

austin_weather = pd.read_csv(url, parse_dates=True, index_col='Date')
austin_weather.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd

def create_scatter_plot(data=None, x=None, y=None, color=None):
    if data is not None:
        x = data[x]
        y = data[y]
        color = data[color]

    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, c=color)
    plt.colorbar()
    plt.show()

# Assuming df is your DataFrame
df = pd.DataFrame({
    'TempHighF': [74, 56, 58, 61, 58],
    'TempLowF': [45, 39, 32, 31, 41],
    'HumidityHighPercent': [93, 93, 76, 89, 86]
})

create_scatter_plot(data=df, x='TempHighF', y='TempLowF', color='HumidityHighPercent')
```

```python
n = 50
fig, ax = plt.subplots(2, 1, sharex=True, figsize=(10, 5))

plots = [
    {
        'ax_index': 0,
        'data': ['TempAvgF', 'TempHighF', 'TempLowF'],
        'color': 'r',
        'ylabel': 'Temperature'
    },
    {
        'ax_index': 1,
        'data': ['DewPointAvgF', 'DewPointHighF', 'DewPointLowF'],
        'color': 'b',
        'ylabel': 'Dew Point'
    }
]


for plot in plots:
    for i, data in enumerate(plot['data']):
        linestyle = '-' if i == 0 else '--' if i == 1 else 'dotted'
        ax[plot['ax_index']].plot(austin_weather.index[:n], austin_weather[data][:n].astype(float), color=plot['color'], linestyle=linestyle)
    ax[plot['ax_index']].set_ylabel(plot['ylabel'])

ax[1].set_xlabel('Date (year-month-day)')
ax[0].set_title('Weather in Austin')
ax[1].xaxis.set_tick_params(rotation=90)

ax[0].annotate(
    'Something happened',
    xy=[pd.Timestamp('2014-01-06'),28],
    xytext=(pd.Timestamp('2014-01-09'),22.5),
    arrowprops={'arrowstyle':'->','color':'gray'}
)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url='https://raw.githubusercontent.com/matzim95/ML-datasets/master/weather.csv'
weather=pd.read_csv( url, parse_dates=True, index_col='date' )


weather_grouped = weather.groupby('weather')['precipitation'].agg(['mean', 'std'])

plt.figure(figsize=(10, 6))
plt.bar(weather_grouped.index, weather_grouped['mean'], yerr=weather_grouped['std'])

plt.xlabel('Typ pogody')
plt.ylabel('Średnie opady')
plt.title('Średnie opady według typu pogody z odchyleniem standardowym')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Załaduj dane
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv')

# Ustaw motyw wykresu
sns.set_theme(style="whitegrid")

# Utwórz histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="displacement", color="skyblue", kde=True)
plt.xlabel('Pojemność silnika')
plt.ylabel('Liczebność')
plt.title('Rozkład pojemności silnika')
plt.show()

# Utwórz jointplot
sns.jointplot(data=data, x="displacement", y="mpg", color="skyblue")
plt.xlabel('Pojemność silnika')
plt.ylabel('Liczba mil na galon')
plt.title('Związek pojemności silnika z liczbą mil na galon')
plt.show()
```
